```python
# profanity checker for the proportion of profanity in all meaningful words (which are raw data without stop words like "I, than, that, do")
def profanity_proportion(text):        
    # converting to lowercase, removing URL links, special characters, punctuations...
    text = text.lower()
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('[’“”…]', '', text)     

    # removing the stop-words          
    text_tokens = word_tokenize(text)
    tokens_without_sw = [word for word in text_tokens if not word in stop_words]
    filtered_sentence = (" ").join(tokens_without_sw)
    text = filtered_sentence
    
    # counts only the swear words
    tokens_with_swearword =[word for word in text_tokens if word in swearwords_list]
    filtered_onlyswearword_sentence = (" ").join(tokens_with_swearword)
    text_profanity = filtered_onlyswearword_sentence
    
    return len(text_profanity.split(" "))/len(text.split(" "))
# currently just testing for the top 100, change this line and calculate more
listby_rap["lyrics"][1:100].apply(profanity_proportion)
```
